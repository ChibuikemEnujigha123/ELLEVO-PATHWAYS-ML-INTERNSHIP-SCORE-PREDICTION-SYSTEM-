##### 3.IDENTIFY AND POSSIBLY REMOVE OUTLIERS

```python

numerical_cols = df.select_dtypes(include='number').columns

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[
        (df[col] < lower) |
        (df[col] > upper)
    ]

    print(f"\n{col}")
    print(f"Number of outliers: {outliers.shape[0]}")
    print(outliers[[col]].head())
```

```python
df = df[
    (df[col] >= lower) &
    (df[col] <= upper)
]
```

```python
df.shape
```

```python
# Columns to remove
cols_to_drop = [
    'Peer_Influence',
    'Learning_Disabilities',	
    'Distance_from_Home',
    'Internet_Access',
    'Parental_Involvement',
    'Access_to_Resources',
    'Extracurricular_Activities',
    'Tutoring_Sessions',
    'Family_Income',
    'Teacher_Quality',
    'Gender',
    'School_Type'
]

# Drop columns
df = df.drop(columns=cols_to_drop)

# Check remaining columns
print(df.columns)
```

### IV.Modelling

```python
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

##### Step 1: Separate Target X and Y

```python
print(df.isnull().sum().sum())
```

```python
# Split features and target
x = df.drop('Exam_Score', axis=1)
y = df['Exam_Score']
```

```python
print(x.dtypes)
```

##### Step 2: Encode Categorical Variables

```python
df.shape
```

```python
#Encoding has taken place previously so there's no need to enocode again
x = pd.get_dummies(x, columns=["Motivation_Level", "Parental_Education_Level"], drop_first=True)
```

##### Step 3:DIVIDING DATA INTO TRAINING ,TESTING AND VALIDATION 

```python
x_train, x_temp, y_train, y_temp = train_test_split(x, y, test_size=0.3, random_state=42)
x_val, x_test, y_val,  y_test = train_test_split (x_temp, y_temp, test_size=0.5 , random_state=42)  
```

```python
# Confirm no NaN values
print(f"NaN values remaining: {x.isnull().sum().sum()}")
```

```python
# 1. Define the columns you encoded
cols = list(df.keys())

# 2. Check for missing values per column
print(df[cols].isnull().sum())

# 3. Check if any NaN exists at all
print("Any NaN in encoded columns?", df[cols].isnull().any().any())

# 4. Total count of missing values
print("Total NaNs:", df[cols].isna().sum().sum())

# 5. Show unique values per column (to spot any unmapped values)
for col in cols:
    print(f"{col}: {df[col].unique()}")
```

##### Step 4:Using models to evaluate output.
##### 4a.Linear Regression Model

```python
linear_model= LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
linear_mod_pred = linear_model.predict(x_val)
```

```python
linear_val_mse = mean_squared_error(y_val, linear_mod_pred)
```

```python
linear_val_r2 = r2_score(y_val, linear_mod_pred)
```

```python
# outputing the result of our MSE and its r2

print(f"The output of the Mean Squared Error (MSE) is:", linear_val_mse)
print(f"The output of its corresponding R-squared is:", linear_val_r2)
```

##### 4b.Decision Tree Regression

```python
decision_tree = DecisionTreeRegressor(random_state = 42)
decision_tree.fit(x_train,y_train)
```

```python
dt_mod_pred = decision_tree.predict(x_val)
dt_val_mse = mean_squared_error(y_val, dt_mod_pred)
dt_val_r2 = r2_score(y_val,dt_mod_pred)
```

```python
# Outputing the result of the DTR and its r2

print(f"The output of the Decision Tree Regressor (DTR) is:", dt_val_mse)
print(f"The output of its corresponding R-squared is:", dt_val_r2)
```

##### 4c.Random Forest Regression

```python
random_forest = RandomForestRegressor(random_state = 42)
random_forest.fit(x_train, y_train)
```

```python
rf_mod_pred = random_forest.predict(x_val)
rf_val_mse = mean_squared_error(y_val,rf_mod_pred)
rf_val_r2 = r2_score(y_val, rf_mod_pred)
```

```python
print(f"The output of the Random Forest Regressor is:", rf_val_mse)
print(f"The output of the  Corresponding R-Squared is:", rf_val_r2 )
```

### V. Hyperparameter Tuning

```python
# initialize RandomForestRegressor with default parameters
rf_model = RandomForestRegressor()
```

```python
# Display the default parameters 
(rf_model.get_params())
```

```python
rf_estimators = RandomForestRegressor(n_estimators=300, random_state=42) # i.e increasing the estimator from the previous 100 to 300
rf_estimators.fit(x_train,y_train)
estimators_val_pred = rf_estimators.predict(x_val)
estimators_val_mse = mean_squared_error(y_val,estimators_val_pred)
estimators_val_r2 = r2_score(y_val, estimators_val_pred)

print(f"Random Forest Regressor (n_estimators=300)")
print(f"Validation MSE :{ estimators_val_mse:.2f}") # f means round the numbe to 2 significant figures
print(f"Validation R2 :{estimators_val_r2:.2f}")
```

```python
rf_depth = RandomForestRegressor(n_estimators=300, max_depth=100, random_state=42) # i.e increasing the estimator from the previous 100 to 300
rf_depth.fit(x_train,y_train)
depth_val_pred = rf_depth.predict(x_val)
depth_val_mse = mean_squared_error(y_val, depth_val_pred)
depth_val_r2 = r2_score(y_val, depth_val_pred)

print(f"Random Forest Regressor (n_estimators=300, max_depth+100)")
print(f"Validation MSE :{ depth_val_mse:.2f}") # f means round the numbe to 2 significant figures
print(f"Validation R2 :{depth_val_r2:.2f}")
```
